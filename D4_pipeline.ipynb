```python
from os import X_OK
x=df.drop(['diagnosis','Unnamed: 32'],axis=1)
y=df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train = LabelEncoder().fit_transform(y_train)
y_test = LabelEncoder().fit_transform(y_test)
```

## Np array to Pytorch tensors

```python
x_tr_ten=torch.from_numpy(X_train)
x_te_ten=torch.from_numpy(X_test)
y_tr_ten=torch.from_numpy(y_train)
y_te_ten=torch.from_numpy(y_test)
```

```python
x_tr_ten
```

# Define model

```python
class MySimpleNN():

  def __init__(self, X):

    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_tr_ten * torch.log(y_pred) + (1 - y_tr_ten) * torch.log(1 - y_pred)).mean()
    return loss


```

## IMP parameter

```python

learning_rate=0.1
epochs=10
```

## Training Pipeline

```python
model =MySimpleNN(x_tr_ten)
# define loops
for epoch in range(epochs):
  # forward pass
  y_pred = model.forward(x_tr_ten)


  # calculate loss
  loss = model.loss_function(y_pred, y_tr_ten)

  # backward pass
  loss.backward()


  # update weights
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero grad
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```

# Evaluation

```python
with torch.no_grad():
  y_pred = model.forward(x_te_ten)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_te_ten).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
