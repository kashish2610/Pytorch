```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
```

```python
from os import X_OK
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train
```

```python
y_train = LabelEncoder().fit_transform(y_train)
y_test = LabelEncoder().fit_transform(y_test)
```

```python
y_train
```

## Np array to Pytorch tensors

```python
x_tr_ten=torch.from_numpy(X_train)
x_te_ten=torch.from_numpy(X_test)
y_tr_ten=torch.from_numpy(y_train)
y_te_ten=torch.from_numpy(y_test)
```

```python
y_te_ten
```

```python
from torch.utils.data import Dataset, DataLoader
class CustomDataset(Dataset):
  def __init__(self,features,labels):
    self.features=features
    self.labels=labels
  def __len__(self):
    return len(self.features)
  def __getitem__(self,idx):
    return self.features[idx],self.labels[idx]
```

```python
tr_ds= CustomDataset(x_tr_ten,y_tr_ten)
te_ds=CustomDataset(x_te_ten,y_te_ten)
```

```python
tr_ds[10]
```

```python
tr_loader= DataLoader(tr_ds,batch_size=32,shuffle=True)
te_loader=DataLoader(te_ds,batch_size=32,shuffle=True)
```

# Define model

```python
import torch
import torch.nn as nn

class MySimpleNN(nn.Module):
  def __init__(self,num_features):

    super().__init__() # invoking nn constructor inheritance concept
    self.Linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()
  def forward(self,features):
    out=self.Linear(features)
    out=self.sigmoid(out)
    return out
```

## IMP parameter

```python

learning_rate=0.1
epochs=10
```

```python
## construct model
model =MySimpleNN(x_tr_ten.shape[1])

#define optimizer
optimizer= torch.optim.SGD(model.parameters(),lr=learning_rate)

# define loss fn
loss_fn=nn.BCELoss()

```

## Training Pipeline

```python
for epoch in range(epochs):
  for features, labels in tr_loader:
    # Forward pass
    outputs = model(features.float())
    loss = loss_fn(outputs, labels.view(-1,1).float())

    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

  # Print loss for every epoch
  print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```

# Evaluation

```python
with torch.no_grad():
  y_pred = model.forward(x_te_ten.float())
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_te_ten).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```

```python
# Model evaluation using test_loader
model.eval()  # Set the model to evaluation mode
accuracy_list = []

with torch.no_grad():
    for batch_features, batch_labels in te_loader:
        # Forward pass
        y_pred = model.forward(x_te_ten.float())
        y_pred = (y_pred > 0.5).float()  # Convert probabilities to binary predictions

        # Calculate accuracy for the current batch
        batch_accuracy = (y_pred.view(-1,1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

# Calculate overall accuracy
overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f'Accuracy: {overall_accuracy:.4f}')
```
